Let's setup the environment by importing the libraries we need and running the necessary Jupyter magic so that plots are displayed inline.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Read the dataset into a DataFrame and start exploring the data.

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

Drop rows with missing values.

```python
recent_grads = recent_grads.dropna()
```

## Create scatter plots to explore the following relations:
## Sample Size vs. Median Salaries
From looking at the different sample sizes, the median salary range for larger sample sizes tend to be between 35,000 and 50,000. The salary range for smaller samples varies quite a bit.

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(5,10))
```


## Sample Size vs. Unemployment Rate
With larger sample sizes, the unemployment rate tends stay consistent between 0.05 and 0.10.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size', figsize=(5,10))
```

## Full-Time vs. Median
With the increase in sample sizes of graduates that were full-time students, the salaries tend to stay between 35,000 and 50,000.



```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs Median', figsize=(5,10))
```

##  Share of Female Graduates vs. Unemployment Rate
There does not look to be any relation between the share of female graduates and unemployment.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate', figsize=(5,10))
```

## Male Graduates vs. Median Salary
With the increase in sample sizes of male graduates, the salaries tend to stay between 35,000 and 60,000.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(5,10))
```

## Female Graduates vs. Median Salary
With the increase in sample sizes of female graduates, the salaries tend to stay between 35,000 and 50,000.

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(5,10))
```

##  Create histograms to explore the distributions of the following:
## Sample Size
The histogram shows the distribution of sample sizes drops off significantly after the first 1,000 graduates.

```python
ax = recent_grads['Sample_size'].hist(bins=10, range=(0,5000), grid=False)
ax.set_title('Sample_size')
```

## Median Salary
The median salary looks to be between 30,000 and 40,000.

```python
ax = recent_grads['Median'].hist(bins=10, range=(0,130000), grid=False)
ax.set_title('Median')
```

## Graduates Employed
This shows that there are more certain major categories that have more graduates employed.

```python
ax = recent_grads['Employed'].hist(bins=10, range=(0,400000), grid=False)
ax.set_title('Employed')
```

## Graduates That Were In School Full-Time
There are certain major categories that have more graduates that went to school full-time.

```python
ax = recent_grads['Full_time'].hist(bins=10, range=(0,400000), grid=False)
ax.set_title('Full_time')
```

## Share of Women Graduates
The highest share of women as a total of all major categories is between 0.7 and 0.8.

```python
ax = recent_grads['ShareWomen'].hist(bins=10, grid=False)
ax.set_title('ShareWomen')
```

##  Unemployment Rate
The highest rate of unemployment as a total of all major categories is between 0.6 and 0.8.

```python
ax = recent_grads['Unemployment_rate'].hist(bins=10, grid=False)
ax.set_title('Unemployment_rate')
```

## Male Graduates
This shows that there are certain major categories that had more male graduates.

```python
ax = recent_grads['Men'].hist(bins=10,grid=False)
ax.set_title('Men')
```

```python
ax = recent_grads['Women'].hist(bins=10,grid=False)
ax.set_title('Women')
```
